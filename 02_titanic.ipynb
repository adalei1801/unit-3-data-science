# Titanic Tabulation

The RMS Titanic was ocean liner that sunk on April 15, 1912 after beginning its maiden voyage 5 days earlier. The Titanic struck an iceberg while carrying 2,224 passengers, of which around 1,500 died when the ship sunk.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv("titanic_passengers.csv")
```

## Question 1 - What port did most of the passengers embark from?


```python
port = titanic_df["Embarked"].value_counts()

port.plot.pie(figsize=(7,8), colors=['royalblue','#EF4135','#009A49'], autopct='%1.1f%%', textprops=dict(color='white', fontsize=15))
plt.title("Ports Passengers Embarked From")
plt.legend(["Southampton", "Cherbourg", "Queenstown"])
```

This graph depicts the ports from which the passengers boarded the Titanic. A majority of the passengers (72.4%) embarked from Southampton, United Kingdom. 18.9% of passengers boarded the Titanic from Cherbourg, France and 8.7% left from Queenstown, Ireland (present day Cobh).

## Question 2 - How many passengers traveled on their own?

```python
travel_pass = titanic_df["Parch"].value_counts()

travel_pass.plot.pie(figsize=(7,8), labeldistance=1.1, textprops=dict(color='white', fontsize=15), colors=['#ad0303', '#db7e21', '#adab05', '#027a06', '#02267a', '#4e027a', '#7a025a'])
plt.title("Passengers with Parents/Children")
plt.ylabel("Amount of Parents or Children Aboard")
plt.legend()
```

The graph depicts the amount of passengers travelled with or without parents or children. 76.1% of passengers travelled alone and around 1% travelled with 5 or more parents or children.

## Question 3 - Were passengers with families on board more likely to survive than passengers without ones?

```python
survival = titanic_df[['Parch', 'Survived']].value_counts()

survival.unstack().plot.bar(color=['#009dc4', '#8B0000'], figsize=(10,5))
plt.suptitle("Rate of Passenger Survival with Families\n")
plt.xlabel("Parents/Children On Board")
plt.ylabel("Passengers")
plt.legend(["Survived", "Died"])
```

This graph represents the survival rates in comparison to parents/children on board. Passengers with no parents or children on board survived more than passengers with parents or children. Passengers with two parents or children on board had an equal amount of survivors. No passenger with 4 or 6 children on board died and death rate of passengers with 1 or 3 parents or children was greater than the survival rate.

## Question 4 - How did ticket prices differ between classes?


```python
titanic_df[titanic_df['Pclass'] == 1].plot.box(column='Fare', by='Pclass', vert=False, figsize=(15, 5), fontsize=12)

plt.title("Fare Price in First Class",fontsize=20, fontfamily='Courier')
plt.xlabel("Fare Price", fontsize=15, fontfamily='Courier')
plt.ylabel("Passenger Class", fontsize=15, fontfamily='Courier')

titanic_df[titanic_df['Pclass'] == 2].plot.box(column='Fare', by='Pclass', vert=False, figsize=(15, 5), fontsize=12)

plt.title("Fare Price in Second Class",fontsize=20, fontfamily='Courier')
plt.xlabel("Fare Price", fontsize=15, fontfamily='Courier')
plt.ylabel("Passenger Class", fontsize=15, fontfamily='Courier')

titanic_df[titanic_df['Pclass'] == 3].plot.box(column='Fare', by='Pclass', vert=False, figsize=(15, 5), fontsize=12)

plt.title("Fare Price in Third Class",fontsize=20, fontfamily='Courier')
plt.xlabel("Fare Price", fontsize=15, fontfamily='Courier')
plt.ylabel("Passenger Class", fontsize=15, fontfamily='Courier')
```

The above graphs display the ticket fare in comparison to passenger class. The median prices between classes ranged from less than $10 to around $70. It also displays the outliers in prices, in first class the furthest outlying price is around $500 and in second and third class its around $70.
