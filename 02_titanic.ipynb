```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv("titanic_passengers.csv")
```

## Question - What port did most of the passengers embark from?


```python
# Chart for question
# Make graph pretty

port = titanic_df["Embarked"].value_counts()

port.plot.pie(figsize=(5,5), colors=['royalblue','#EF4135','#009A49'], legend=True)
```

How many passengers traveled on their own?

```python
travel_pass = titanic_df["Parch"].value_counts()

plt.title("Passengers Family Count")
travel_pass.plot.pie(figsize=(5,5), legend=True, labeldistance=1.1, autopct='%1.1f%%')
```
